# 02 - Translation Test
## Test Translation Quality Before Full Dataset Translation

This notebook tests all 3 translation variants on a small sample:
1. **Pure Urdu** - Full Urdu script
2. **Urdu-English Mixed** - Urdu script with some English words
3. **Roman Urdu** - Urdu written in English letters

```python
import sys
sys.path.insert(0, '..')

import pandas as pd
from src.utils import INTERIM_DIR, FRIENDLY_LABELS
from src.translator import (
    translate_to_urdu,
    create_mixed_variant,
    transliterate_to_roman,
)
```

## Step 1: Load a Small Sample from Interim Data

```python
# Load mapped training data
train_df = pd.read_csv(INTERIM_DIR / 'train_mapped.csv', encoding='utf-8-sig')
print(f'Total rows: {len(train_df):,}')

# Take 10 samples (mix of all labels)
sample = train_df.groupby('label_text').head(4).head(12).reset_index(drop=True)
print(f'Sample size: {len(sample)}')
sample[['claim', 'label_text']].head(12)
```

## Step 2: Test Pure Urdu Translation

```python
# Test Pure Urdu translation on sample claims
print('PURE URDU TRANSLATION TEST')
print('=' * 70)

for i, row in sample.iterrows():
    english = row['claim']
    urdu = translate_to_urdu(english)
    
    print(f'\n[{i+1}] Label: {row["label_text"]}')
    print(f'    EN:   {english}')
    print(f'    URDU: {urdu}')
    print('-' * 70)
```

## Step 3: Test Urdu-English Mixed

```python
# Test Mixed variant
print('URDU-ENGLISH MIXED TEST')
print('=' * 70)

for i, row in sample.head(5).iterrows():
    english = row['claim']
    urdu = translate_to_urdu(english)
    mixed = create_mixed_variant(english, urdu)
    
    print(f'\n[{i+1}]')
    print(f'    EN:    {english}')
    print(f'    URDU:  {urdu}')
    print(f'    MIXED: {mixed}')
    print('-' * 70)
```

## Step 4: Test Roman Urdu Transliteration

```python
# Test Roman Urdu variant
print('ROMAN URDU TRANSLITERATION TEST')
print('=' * 70)

for i, row in sample.head(5).iterrows():
    english = row['claim']
    urdu = translate_to_urdu(english)
    roman = transliterate_to_roman(urdu)
    
    print(f'\n[{i+1}]')
    print(f'    EN:    {english}')
    print(f'    URDU:  {urdu}')
    print(f'    ROMAN: {roman}')
    print('-' * 70)
```

## Step 5: Side-by-Side Comparison (All 3 Variants)

```python
# Full comparison table
results = []
for i, row in sample.head(8).iterrows():
    english = row['claim']
    urdu = translate_to_urdu(english)
    mixed = create_mixed_variant(english, urdu)
    roman = transliterate_to_roman(urdu)
    
    results.append({
        'English': english[:60],
        'Pure Urdu': urdu[:60],
        'Mixed': mixed[:60],
        'Roman Urdu': roman[:60],
        'Label': row['label_text'],
    })

comparison_df = pd.DataFrame(results)
comparison_df.style.set_properties(**{'text-align': 'left'})
```
